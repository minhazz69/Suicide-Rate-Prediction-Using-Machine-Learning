# Data Pre-processing

```python
#initial necessary libraries
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#check null values
df.isnull().sum() 
```

```python
#remove unecessary coloumns
df.drop(["hdi_for_year"],axis=1,inplace=True)
df.drop(["country_year"],axis=1,inplace=True) 
```

```python
df.isnull().sum()
```

```python
#remove null row values
#drop null values
df_clean = df.dropna() 
df_clean.isnull().sum()
```

```python
df_clean.head()
```

## ONE HAT ENCODING

```python
#ONE HAT ENCODING (replacing string values to binary values)
from sklearn.preprocessing import OneHotEncoder
```

```python
df_clean['sex'].unique()
```

```python
df_clean['age'].unique()
```

```python
df_clean['generation'].unique()
```

```python
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output = False).set_output(transform='pandas')
```

```python
ohe_sex = ohe.fit_transform(df_clean[['sex']])
```

```python
ohe_age = ohe.fit_transform(df_clean[['age']])
```

```python
ohe_generation = ohe.fit_transform(df_clean[['generation']])
```

```python
ohe_sex.head()
```

```python
ohe_age.head()
```

```python
ohe_generation.head()
```

```python
df_mod = pd.concat([df_clean,ohe_sex,ohe_age,ohe_generation], axis=1).drop(columns=['generation','sex','age'])
```

```python
df_mod.head()
#figure7
```

```python
df_mod.shape
```

# Random Forest Model

```python
#dataset split to X and Y for Random Forest Model 
x=df_mod.drop(['ï»¿country','gdp_for_year'],axis=1)
y=df_mod['suicides_no']
```

```python
#training dataset=80% and testing dataset=20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x, y, test_size=0.20)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
#RF implement
rf_model=RandomForestRegressor(n_estimators=1500, random_state=100) #hyper parameters
value_rf=rf_model.fit(x_train,y_train)
ytrain_predict=value_rf.predict(x_train)
ytest_predict=value_rf.predict(x_test)
```

```python
#Model Evaluation using R2,MAE,MSE,RMSE
print('Random Forest training data default hyperparameter R-squared, R2=%.2f'% r2_score(y_train,ytrain_predict))
print('Random Forest test data default hyperparameter R-squared,     R2=%.2f \n'% r2_score(y_test,ytest_predict))
```

```python
print('Random Forest training data default hyperparameter Mean-Absolute-Error, MAE= %.2f'% mean_absolute_error(y_train,ytrain_predict))
print('Random Forest test data default hyperparameter Mean-Absolute-Error,     MAE= %.2f \n'% mean_absolute_error(y_test,ytest_predict))
```

```python
print('Random Forest training data default hyperparameter Mean-Squared-Error, MSE= %.2f'% mean_squared_error(y_train,ytrain_predict))
print('Random Forest test data default hyperparameter Mean-Squared-Error,     MSE= %.2f \n'% mean_squared_error(y_test,ytest_predict))
```

```python
print('Random Forest training data default hyperparameter Root-Mean-Squared-Error, RMSE= %.2f'% np.sqrt(mean_squared_error(y_train,ytrain_predict)))
print('Random Forest test data default hyperparameter Root-Mean-Squared-Error,     RMSE= %.2f'% np.sqrt(mean_squared_error(y_test,ytest_predict)))
```

```python
#Random Forest visualization
plt.rcParams['figure.figsize']=(20,20) #VAR
x_ax=range(len(x_test))
plt.plot(x_ax,y_test,label='observed', color='blue', linestyle='-')    #training
plt.plot(x_ax,ytest_predict,label='predicted', color='red', linestyle='--') #test
plt.ylabel('suicide no')
plt.xlabel('test sample')
plt.legend(bbox_to_anchor=(0.5,-0.2), loc='lower center', ncol=2, frameon=False)
plt.show()
#figure8
```

# Support Vector Regressor(SVR)

```python
from sklearn.svm import SVR
```

```python
#dataset split to X and Y for Support Vector Regressor model 
x=df_mod.drop(['ï»¿country','gdp_for_year'],axis=1)
y=df_mod['suicides_no']
```

```python
#training dataset=80% and testing dataset=20%
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x, y, test_size=0.20)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
##PLAY WITH  c and epsilon
model_svr = SVR(kernel='rbf', C=20, epsilon=0.3) #hyper parameter
value_svr = model_svr.fit(x_train, y_train)
ytrain_predict=value_svr.predict(x_train)
ytest_predict=value_svr.predict(x_test)
```

```python
#Model Evaluation using R2,MAE,MSE,RMSE
print('SVR training data default hyperparameter R-squared, R2=%.2f'% r2_score(y_train,ytrain_predict))
print('SVR test data default hyperparameter R-squared,     R2=%.2f \n'% r2_score(y_test,ytest_predict))
```

```python
print('SVR training data default hyperparameter Mean-Absolute-Error, MAE= %.2f'% mean_absolute_error(y_train,ytrain_predict))
print('SVR test data default hyperparameter Mean-Absolute-Error,     MAE= %.2f \n'% mean_absolute_error(y_test,ytest_predict))
```

```python
print('SVR training data default hyperparameter Mean-Squared-Error, MSE= %.2f'% mean_squared_error(y_train,ytrain_predict))
print('SVR test data default hyperparameter Mean-Squared-Error,     MSE= %.2f \n'% mean_squared_error(y_test,ytest_predict))
```

```python
print('SVR training data default hyperparameter Root-Mean-Squared-Error, RMSE= %.2f'% np.sqrt(mean_squared_error(y_train,ytrain_predict)))
print('SVR test data default hyperparameter Root-Mean-Squared-Error,     RMSE= %.2f'% np.sqrt(mean_squared_error(y_test,ytest_predict)))
```
